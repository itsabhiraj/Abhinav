```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
x,y = make_regression(n_samples= 4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=10)
```

```python
import matplotlib.pyplot as plt
print(y)
plt.scatter(x,y)
```

```python
lr = LinearRegression()
lr.fit(x,y)
lr.coef_, lr.intercept_

# slope, b_original
```

```python
plt.plot(x, lr.predict(x), c='r')
plt.scatter(x,y)
```

```python
#gradient descent, let's assume b=10
#pred = ((lr.coef_ * x )+ b)
b=10
pred = ((lr.coef_ * x )+ b)
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x), c='r', label='ols')
plt.plot(x,pred, label = 'b=10')
plt.legend()
plt.show()
```

```python
#loss_slope = -2*np.sum(y-m*x.ravel() - b)
loss_slope = -2*np.sum(y-(lr.coef_)*x.ravel() - b)
loss_slope
```

```python
# b_new = b_old - losss_slope*learning_rate
#let the learning rate be 0.1
l_r=0.1
b = b - l_r*loss_slope
b
```

```python
#we will run this in a loop to reach our desired output

epoch=5
for i in range(epoch):
    loss_slope = -2*np.sum(y-(lr.coef_)*x.ravel() - b)
    b = b - l_r*loss_slope
    pred_i = lr.coef_*x +b
    plt.plot(x,pred_i, c='b')
plt.scatter(x,y)
plt.plot(x,lr.predict(x), c='y', label = 'ols')
plt.plot(x,pred, label ='b=10')
plt.legend()
plt.show()
```
